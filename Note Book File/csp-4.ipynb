```python

# create pivot table
data=pd.DataFrame({"Gender":["Male","Female"],
                   "count":[round((177/180)*100,2),round((3/180)*100,2)]})
pivot_table=pd.pivot_table(data,index="Gender",values="count",columns=None)
pivot_table
#draw pie chart
labels=pivot_table.index
values=pivot_table["count"]
fig,ax=plt.subplots(figsize=(10,6),subplot_kw=dict(aspect="equal"))
def func(pct,allvals):
  absolute=int(np.round(pct/100.*np.sum(allvals)))
  return f"{pct:.1f}%\n({absolute:d}%)"
wedges, texts, autotexts=ax.pie(values, autopct=lambda pct: func(pct,values),
                                textprops=dict(color="b"),colors=["red","pink"],
                                shadow=True,startangle=0,explode=[0.6,0],
                                frame=True)
legend=ax.legend(wedges,labels,title="Gender",loc="center left",
                 bbox_to_anchor=(1,0,0.5,1),shadow=True,fontsize="x-large")
legend.get_frame().set_facecolor("C2")
plt.setp(autotexts,size=8,weight="bold")
ax.set_facecolor("seashell")
ax.set_title("Gender")
plt.show()
```

```python
df["ADDRESS"].value_counts()
```

```python
data={"ADDRESS":df["ADDRESS"],
      "Age":df["Age"]}
df1=pd.DataFrame(data)
df1
#define age ranges
age_ranges={"25 Below":(0,25),
            "26-35":(26,35),
            "36-45":(36,45),
            "46-55":(46,55),
            "56-65":(56,65),
            "66 Above":(66,float("inf"))}
filtered_data={}
uniquee_address=df["ADDRESS"].unique()
for address in uniquee_address:
  address_data=df[df["ADDRESS"]==address]
  #intialize the range counters
  age_range_counts={key:0 for key in age_ranges}
  #count ages in different ranges
  for index , row in address_data.iterrows():
    age=row["Age"]
    for range_name,(lower,upper) in age_ranges.items():
      if lower<=age <=upper:
        age_range_counts[range_name]+=1
        #store the results for this address
        filtered_data[address]=age_range_counts
for address ,age_range_counts in filtered_data.items():
  print("ADDRESS:",address)
  for range_name,count in age_range_counts.items():
    print(f"{range_name}:{count}")
  print()
```
